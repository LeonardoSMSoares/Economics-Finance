**3.1 - An article in the journal Air and Waste (Update on Ozone Trends in
California’s South Coast Air Basin, Vol. 43, 1993) investigated the
ozone levels in the South Coast Air Basin of California for the years
1976–1991. The author believes that the number of days the ozone
levels exceeded 0.20 ppm (the response) depends on the seasonal
meteorological index, which is the seasonal average 850-millibar
Temperature (the predictor). Table E3.1 gives the data.**

```python
# Define the data
data = {
    'Year': [1976, 1977, 1978, 1979, 1980, 1981, 1982, 1983, 1984, 1985, 1986, 1987, 1988, 1989, 1990, 1991],
    'Days': [91, 105, 106, 108, 88, 91, 58, 82, 81, 65, 61, 48, 61, 43, 33, 36],
    'Index': [16.7, 17.1, 18.2, 18.1, 17.2, 18.2, 16.0, 17.2, 18.0, 17.2, 16.9, 17.1, 18.2, 17.3, 17.5, 16.6]
}
# Create DataFrame
df_days_index = pd.DataFrame(data)

# Show the DataFrame
df_days_index
```

**a. Construct a scatter diagram of the data.**

```python
# Scatter plot of Days vs Index
plt.figure(figsize=(8, 5))
plt.scatter(df_days_index['Index'], df_days_index['Days'], color='blue', edgecolor='black')
plt.title('Days that Ozone Levels Exceed 20 ppm')            
plt.xlabel('Days')                                       
plt.ylabel('Index')                                      
plt.grid(True, linestyle='--', alpha=0.6)                 
plt.tight_layout()
plt.show()
```

**b. Estimate the prediction equation.**

```python
# Define X and Y
X = df_days_index['Index']                      # Predictor
Y = df_days_index['Days']                       # Response

# Add constant (for intercept)
X_with_const = sm.add_constant(X)

# Fit OLS model
model = sm.OLS(Y, X_with_const).fit()

# Display summary
print(model.summary())
```

```python
# Create prediction line
x_vals = np.linspace(X.min(), X.max(), 100)              # Generate X values
x_vals_const = sm.add_constant(x_vals)                   # Add intercept term
y_pred = model.predict(x_vals_const)                     # Predict Y

# Plot
plt.figure(figsize=(8, 5))
plt.scatter(X, Y, color='blue', label='Observed Data')   # Scatter plot of original data
plt.plot(x_vals, y_pred, color='red', label='Regression Line')  # Regression line
plt.xlabel('Index (Seasonal Meteorological Index)')
plt.ylabel('Days (Ozone > 0.20 ppm)')
plt.title('Ozone Exceedance vs. Meteorological Index')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

**d. Calculate the 95% CI and PI on for a seasonal meteorological

**f. Is there any evidence of autocorrelation in the residuals?**

```python
residuals = model.resid                         # Extract residuals from OLS model
plot_acf(residuals, lags=10)                    # Plot autocorrelation up to 10 lags
plt.title('ACF of Regression Residuals')
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.grid(True)
plt.show()
```

```python
# Calculate the Durbin-Watson statistic (from statsmodels.stats.stattools import durbin_watson)
dw_stat = durbin_watson(residuals)
print(f"Durbin-Watson statistic: {dw_stat:.4f}")

# Interpretation based on DW value
if abs(dw_stat - 2) < 0.15:
    interpretation = "No autocorrelation"
elif dw_stat < 2:
    interpretation = "Positive autocorrelation"
elif dw_stat > 2:
    interpretation = "Negative autocorrelation"
else:
    interpretation = "Strong autocorrelation (positive or negative)"

print(f"Interpretation: {interpretation}")
```

**3.2 - Montgomery, Peck, and Vining (2012) present data on the number
of pounds of steam used per month at a plant. Steam usage is thought
to be related to the average monthly ambient temperature. The past
year’s usages and temperatures are shown in Table E3.2.**

```python
# Data from Table E3.2
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Temperature': [21, 24, 32, 47, 50, 59, 68, 74, 62, 50, 41, 30],
    'SteamUsage': [185.79, 214.47, 288.03, 424.84, 454.68, 539.03, 621.55, 675.06, 562.03, 452.93, 369.95, 273.98]
}

df = pd.DataFrame(data)

df.head()
```

```python
# Define predictor (X) and response (y)
X = sm.add_constant(df['Temperature'])  # Add intercept
y = df['SteamUsage']

# Fit the OLS regression model
model = sm.OLS(y, X).fit()

# Display regression summary
print(model.summary())
```

**b. Test for significance of regression.**

```python
# Display regression summary
print(model.summary())
```

**d. Plant management believes that an increase in average ambient
temperature of one degree will increase average monthly steam
consumption by 10,000 lb. Do the data support this statement?**

#### Testing the Claim about Temperature Increase Impact

**Claim:** A 1°F increase in ambient temperature increases steam usage by **10,000 lb**, which is **10 units** (in 1,000 lb).

We look at the estimated **slope coefficient** from the regression model:

- **Estimated Slope (β₁)**: 9.2085
- **Standard Error of Slope**: 0.034
- **Hypothesized Value**: 10

---

#### Hypothesis Test

- **Null Hypothesis ($H_0$):** $ \beta_1 = 10 $
- **Alternative Hypothesis ($H_1$):** $ \beta_1 \neq 10 $

We compute the test statistic:

$$
t = \frac{\hat{\beta}_1 - 10}{SE(\hat{\beta}_1)} = \frac{9.2085 - 10}{0.034} \approx −23.28
$$

Degrees of freedom: $n - 2 = 12 - 2 = 10$

Using a t-distribution, the **critical t-value** at 5% significance level (two-tailed) ≈ **-23.28**.

The p-value is much less than 0.05, meaning **we reject the null hypothesis**.

---

#### Conclusion:

At the 5% significance level, *the data do not support the plant management's claim that each 1°F increase causes a 10,000 lb increase in steam usage*.
Instead, the actual estimated increase is 9,208.5 lb, which is significantly less than 10,000 lb.

**b. Use the Durbin–Watson test to determine if there is positive
autocorrelation in the errors. What are your conclusions?**

The Durbin-Watson Statistic (DWS) is a diagnostic tool used to detect autocorrelation in the residuals of a regression model — particularly first-order autocorrelation, which occurs when the residual at time $t$ is correlated with the residual at time $t-1$.

```python
# Calculate the Durbin-Watson statistic (from statsmodels.stats.stattools import durbin_watson)
dw_stat = durbin_watson(residuals)
print(f"Durbin-Watson statistic: {dw_stat:.4f}")

# Interpretation based on DW value
if abs(dw_stat - 2) < 0.15:
    interpretation = "No autocorrelation"
elif dw_stat < 2:
    interpretation = "Positive autocorrelation"
elif dw_stat > 2:
    interpretation = "Negative autocorrelation"
else:
    interpretation = "Strong autocorrelation (positive or negative)"

print(f"Interpretation: {interpretation}")
```

**c. Use one iteration of the Cochrane–Orcutt procedure to estimate
the regression coefficients. Find the standard errors of these
regression coefficients.**

##### 1. Fit the Initial OLS Model

**Model:**

$$
y_t = \beta_0 + \beta_1 x_t + \varepsilon_t
$$

```python
# Step 1: Define predictors and response
X = sm.add_constant(df_peaches['Price (x_t)'])     # Add constant term
y = df_peaches['MarketShare (y_t)']                # Response variable

# Fit the regression model
model = sm.OLS(y, X).fit()

residuals = model.resid                            # Get residuals
```

##### 2. Estimate $\rho$ (Autocorrelation Coefficient)

**Model:**

$$
e_t = \rho e_{t-1} + u_t
$$


```python
# Step 2: Estimate ρ using lagged residuals
resid_lag = residuals.shift(1)[1:]            # e_{t-1}
resid_current = residuals[1:]                 # e_t

rho_model = sm.OLS(resid_current, sm.add_constant(resid_lag)).fit()
rho = rho_model.params.iloc[1]                # Slope coefficient is ρ
rho
```

##### 3. Transform the Variables

**Transformation Equations:**

$$
y_t^* = y_t - \rho y_{t-1}, \quad x_t^* = x_t - \rho x_{t-1}
$$


```python
# Step 3: Transform the variables
y_star = y[1:] - rho * y.shift(1)[1:]
x_star = df_peaches['Price (x_t)'][1:] - rho * df_peaches['Price (x_t)'].shift(1)[1:]
X_star = sm.add_constant(x_star)              # Add constant for intercept
X_star.head()
```

```python
# Step 4: Fit transformed regression model
orcutt_model = sm.OLS(y_star, X_star).fit()
print(orcutt_model.summary())
```

**Transformed Regression Equation:**

$$
y_t^* = \beta_0 (1 - \rho) + \beta_1 x_t^* + u_t

### Iterative Steps of the Cochrane–Orcutt Procedure

1. **Estimate the model via OLS:**

$$
y_t = \beta_0 + \beta_1 x_t + \varepsilon_t
$$

2. **Estimate the autocorrelation coefficient** \( \rho \) **from the residuals:**

$$
e_t = \rho e_{t-1} + u_t
$$

3. **Transform the data using** \( \rho \):

$$
y_t^* = y_t - \rho y_{t-1}
$$

$$
x_t^* = x_t - \rho x_{t-1}
$$

4. **Fit the regression to the transformed variables:**

$$
y_t^* = \beta_0 (1 - \rho) + \beta_1 x_t^* + u_t
$$

5. **Repeat steps 2–4 using residuals from the transformed model to update** $ \rho $ **until convergence** *(i.e., when the change in* $ \rho $ *becomes negligible)*.


```python
# Define initial X and y
X = sm.add_constant(df_peaches['Price (x_t)'])     # Predictor with intercept
y = df_peaches['MarketShare (y_t)']                # Response variable

# Step 1: Fit initial OLS model
model_ols = sm.OLS(y, X).fit()
resid = model_ols.resid

# Step 2: Initialize parameters
rho_old = 0
tolerance = 1e-5
max_iter = 100

for iteration in range(max_iter):
    
    # Estimate new ρ using regression of e_t on e_{t-1}
    resid_lag = resid.shift(1)[1:]
    resid_current = resid[1:]
    resid_lag.name = 'Residuals_lag'  # Explicit name for the predictor
    rho_model = sm.OLS(resid_current, sm.add_constant(resid_lag)).fit()
    rho_new = rho_model.params['Residuals_lag']
    
    # Stop if convergence is reached
    if np.abs(rho_new - rho_old) < tolerance:
        break
    rho_old = rho_new

    # Step 3: Transform variables
    y_star = y[1:] - rho_new * y.shift(1)[1:]
    x_star = df_peaches['Price (x_t)'][1:] - rho_new * df_peaches['Price (x_t)'].shift(1)[1:]
    X_star = sm.add_constant(x_star)

    # Step 4: Fit transformed model
    model_transformed = sm.OLS(y_star, X_star).fit()
    resid = model_transformed.resid  # Update residuals

# Final Cochrane–Orcutt model output
print(f"Converged after {iteration+1} iterations.")
print(f"Estimated ρ: {rho_new:.4f}")
print(model_transformed.summary())
```
