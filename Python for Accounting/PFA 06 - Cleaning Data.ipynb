### Discarding missing values

```python
ledger_df.head()
```

```python
ledger_df[ledger_df['Product Name'].notna()].head()
```

```python
ledger_df['Product Name'].notna()
```

```python
ledger_df.dropna().head()
```

```python
ledger_df.head()
```

```python
ledger_df.dropna(subset=['Product Name', 'ProductID']).head()
```

```python
ledger_df.dropna(subset=['AccountNo', 'ProductID']).head()
```

This example would discard only those rows from `ledger_df` that are entirely empty (i.e., values in all columns are NaN)

```python
ledger_df.dropna(how='all').head()
```

Instead of rows, you can also tell dropna to **discard columns** with
empty values instead, using the axis keyword argument. To discard
all columns in `ledger_df` that have any empty values in them, you
can use:

```python
ledger_df.dropna(how='any', axis='columns')   # In this case, 'Product Name' will be kick out.
```

### Filling missing values

```python
# Replace any missing (NaN) values in the 'Product Name' column with the string 'MISSING'
ledger_df['Product Name'].fillna('MISSING')  
```

```python
ledger_df.head()
```

```python
# Perform forward fill to replace missing values with the previous non-null value
ledger_df['Product Name'] = ledger_df['Product Name'].ffill()  

ledger_df.head()
```

### Dealing with duplicate rows

```python
ledger_df.duplicated()
```

```python
ledger_df.loc[3] == ledger_df.loc[4]
```

```python
# Filter out duplicate rows from 'ledger_df' by selecting only rows that are not duplicated
ledger_df[~ledger_df.duplicated()].head()
```

```python
ledger_df.drop_duplicates()

ledger_df.head()
```

You get the same result as above, but with less typing. You can also
tell pandas to check for duplicates in a subset of columns only, by
passing a list of column names to the subset keyword argument:

```python
# Remove duplicate rows based on 'InvoiceNo' and 'ProductID' columns
ledger_df.drop_duplicates(subset=['InvoiceNo', 'ProductID']).head()  
```

### Converting column data types

```python
# Read Dataframe 'Q1Sales.csv'

url = ("https://raw.githubusercontent.com/pythonforaccounting/workspace/refs/heads/main/P2%20-%20Working%20with%20tables/Q1Sales.csv")
ledger_df = pd.read_csv(url)

ledger_df.head()
```

```python
ledger_df.info()
```

Another useful DataFrame method that can fix issues with column
data types is `convert_dtypes`. This method asks pandas to try and
figure out better data types for each column in a DataFrame, and
can often reduce the amount of computer memory needed to store
your data:

```python
ledger_df.convert_dtypes().info()
```

The `convert_dtypes` method also replaces all missing values in the
DataFrame (i.e., any NaN, NaT or None values) with the unified <NA>
sentinel value:

```python
ledger_df.convert_dtypes()
```

The changes produced by calling `convert_dtypes` are not critical, and you can still work with ledger_df without using `con
vert_dtypes`. However, calling `convert_dtypes` on a DataFrame can
make it use less computer memory, and for large tables, that can
speed up other table operations later on.
