```python
# Direct form:

sales_df = pd.concat([  # Concatenate the data from multiple Excel sheets into a single DataFrame
        pd.read_excel('data/Q1Sales.xlsx', sheet_name='January'),   # Read data from the 'January' sheet
        pd.read_excel('data/Q1Sales.xlsx', sheet_name='February'),  # Read data from the 'February' sheet
        pd.read_excel('data/Q1Sales.xlsx', sheet_name='March')      # Read data from the 'March' sheet
    ], 
    ignore_index=True)  # Ignore the original row indices and reset the index in the concatenated DataFrame

sales_df
```

```python
channel = 'Understock.com'
channel_df = sales_df[sales_df['Channel'] == channel]
channel_df = channel_df.sort_values(['Total', 'Quantity'], ascending=False)
```

```python
# Define the channel name to filter data from the 'sales_df' DataFrame
channel = 'Understock.com'  

# Filter the 'sales_df' DataFrame to include only rows where the 'Channel' column is equal to 'Understock.com'
channel_df = sales_df[sales_df['Channel'] == channel]  

# Sort the filtered data first by 'Total' in descending order, and then by 'Quantity' in descending order
channel_df = channel_df.sort_values(['Total', 'Quantity'], ascending=False)  

channel_df
```

We can write `channel_df` to a new Excel file using:

```python
channel_df.to_excel('data/Q1ChannelSales.xlsx', sheet_name='Understock.com', index=False)
```

```python
channels = ['Bullseye', 'iBay.com', 'Shoppe.com', 'Understock.com', 'Walcart']

for channel in channels:
    channel_df = sales_df[sales_df['Channel'] == channel]
    channel_df = channel_df.sort_values(['Total', 'Quantity'], ascending=False)

channel_df
```

Why 'Walcart' sales? Because the for loop goes through all values
in the channels list and keeps assigning a different DataFrame to
channel_df. The last value in the channels list is 'Walcart', so at
the last iteration of the loop, channel_df gets assigned only those
rows in sales_df where values in the 'Channel' column are equal
to 'Walcart'.

```python
# Define a list of channel names to iterate over
channels = ['Bullseye', 'iBay.com', 'Shoppe.com', 'Understock.com', 'Walcart']  

# Create an ExcelWriter object to write data to a new Excel file 'Q1ChannelSales.xlsx'
output_file = pd.ExcelWriter('data/Q1ChannelSales.xlsx')

for channel in channels:                                                         # Loop through each channel in the 'channels' list
    channel_df = sales_df[sales_df['Channel'] == channel]                        # Filter the 'sales_df' with only rows for the current channel
    channel_df = channel_df.sort_values(['Total', 'Quantity'], ascending=False)  # Sort the filtered data by 'Total' and 'Quantity' in descending order
    channel_df.to_excel(output_file, sheet_name=channel, index=False)            # Write the sorted data to a new sheet in the Excel file

output_file.close()                                                              # Save the Excel file after all sheets have been written
```

Another way:

```python
# Define a list of channel names to iterate over
channels = ['Bullseye', 'iBay.com', 'Shoppe.com', 'Understock.com', 'Walcart']  

with pd.ExcelWriter('data/Q1ChannelSales2.xlsx') as output_file:
    for channel in channels:
        channel_df = sales_df[sales_df['Channel'] == channel]
        channel_df = channel_df.sort_values(['Total', 'Quantity'], ascending=False)
        channel_df.to_excel(output_file, sheet_name=channel, index=False)
```

Concatenating all:

```python
sales_df = pd.concat([
        pd.read_excel('data/Q1Sales.xlsx', sheet_name='January'), 
        pd.read_excel('data/Q1Sales.xlsx', sheet_name='February'), 
        pd.read_excel('data/Q1Sales.xlsx', sheet_name='March')
    ], ignore_index=True)

channels = ['Bullseye', 'iBay.com', 'Shoppe.com', 'Understock.com', 'Walcart']

with pd.ExcelWriter('data/Q1ChannelSales4.xlsx') as output_file:
    for channel in channels:
        channel_df = sales_df[sales_df['Channel'] == channel]
        channel_df = channel_df.sort_values(['Total', 'Quantity'], ascending=False)
        channel_df.to_excel(output_file, sheet_name=channel, index=False)
```
