# Chapter 17 - Adding and Modifying Columns

```python
import pandas as pd
```

### Data Set

```python
# Read Dataframe 'Q1Sales.csv'

url = ("https://raw.githubusercontent.com/pythonforaccounting/workspace/refs/heads/main/P2%20-%20Working%20with%20tables/Q1Sales.csv")
ledger_df = pd.read_csv(url)

ledger_df.head()
```

### Adding Columns

```python
ledger_df['Quarter'] = 'Q1'

ledger_df.head()
```

```python
ledger_df['Quarter'] = 23

ledger_df.head()
```

```python
ledger_df['Total'] * (19 / 100)
```

```python
ledger_df['Sales Tax'] = ledger_df['Total'] * (19 / 100)

ledger_df.head()
```

```python
ledger_df['Quantity'] * ledger_df['Unit Price']
```

```python
ledger_df['Total'] = ledger_df['Quantity'] * ledger_df['Unit Price']

ledger_df.head()
```

### Renaming Columns

```python
ledger_df.rename(columns={"Sales Tax": "Tax"}, inplace=True)

ledger_df.head()
```

```python
ledger_df.columns
```

```python
# Renaming multiple columns
ledger_df = ledger_df.rename(columns={
    'Product Name': 'Name',
    'Unit Price': 'Price',
    'Hello': 'World'
})

ledger_df.head()
```

### Replacing Values

```python
# Perform the replacement directly on the 'Channel' column
ledger_df['Channel'] = ledger_df['Channel'].replace({
    'Shoppe.com': 'Shoppe',
    'Understock.com': 'Understock',
    'iBay.com': 'iBay'
})

ledger_df.head()
```

### Assigning new values to a table slice

```python
ledger_df.loc[1, 'ProductID']
```

```python
# with .loc
ledger_df.loc[1, 'ProductID'] = 'T&G/LEG0-0190'

ledger_df.head()
```

```python
# or with .iloc
ledger_df.iloc[2, 3] = 'T&G/PET-01099'

ledger_df.head()
```

```python
# Convert the 'Quarter' column to string type if it isn't already
ledger_df['Quarter'] = ledger_df['Quarter'].astype(str)

# Now assign the value 'One' to the first three rows in the 'Quarter' column
ledger_df.loc[0:2, 'Quarter'] = 'One'

ledger_df.head()
```

```python
ledger_df.loc[0:2, 'Quarter'] = ['One', 'Two', 'Three']

ledger_df.head()
```

### Filter and edit

```python
ledger_df.loc[ledger_df['Channel'] == 'Walcart', 'Total']
```

To calculate the new tax amount, you can use:

```python
ledger_df.loc[ledger_df['Channel'] == 'Walcart', 'Total'] * (9 / 100)
```

Another way:

```python
walcart_rows = ledger_df['Channel'] == 'Walcart'

walcart_tax = ledger_df.loc[walcart_rows, 'Tax'] = ledger_df.loc[walcart_rows, 'Total'] * (9 / 100)

walcart_tax
```

```python
# Filters and returns rows where the 'Channel' column equals 'Walcart'
ledger_df[ledger_df['Channel'] == 'Walcart'].head()
```

```python
ledger_df.loc[ledger_df['Channel'] == 'Walcart', 'Tax'] = (
    ledger_df.loc[ledger_df['Channel'] == 'Walcart', 'Total'] * (8 / 100)   # Before was 9/100
)

ledger_df[ledger_df['Channel'] == 'Walcart'].head()
```

```python
ledger_df[ledger_df['Channel'] == 'Bullseye'].head()
```

```python
ledger_df.loc[ledger_df['Channel'] == 'Bullseye', 'Tax'] = (
    ledger_df.loc[ledger_df['Channel'] == 'Bullseye', 'Total'] * (16 / 100)
)

ledger_df[ledger_df['Channel'] == 'Bullseye'].head()
```
