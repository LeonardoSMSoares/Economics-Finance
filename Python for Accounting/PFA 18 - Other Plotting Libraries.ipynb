# Chapter 32 - Other Plotting Libraries

```python
import pandas as pd
```

### Data set

```python
daily_sales_df = pd.read_csv('data/Q1DailySales.csv', parse_dates=['Date'])

daily_sales_df = daily_sales_df.set_index('Date')

daily_sales_df
```

### Plotting with pandas

```python
daily_sales_df.plot();
```

```python
ax = daily_sales_df.plot()

ax.legend(bbox_to_anchor=(1, 1), loc='upper left');
```

### Different kinds of pandas plots

```python
daily_sales_df.plot(kind='line');
```

```python
daily_sales_df.plot(
    kind='area',
    legend='reverse'
);
```

```python
daily_sales_df.sum().plot(
    kind='pie',
    explode=[0, 0, 0.1, 0, 0]
);
```

```python
daily_sales_df.sum().plot(
    kind='bar',
    rot=20
);
```

```python
daily_sales_df.sum().plot(kind='barh');
```

```python
daily_sales_df.plot(
    kind='hist',
    bins=20
);
```

```python
daily_sales_df.plot(kind='density');
```

```python
daily_sales_df.plot(
    kind='box',
    rot=30
);
```

```python
daily_sales_df.plot(
    kind='scatter',
    x='iBay.com',
    y='Understock.com'
);
```

```python
daily_sales_df.plot(
    kind='hexbin',
    x='iBay.com',
    y='Understock.com',
    gridsize=10,
    sharex=False
);
```

```python
(daily_sales_df
    .plot
    .scatter(
        x='iBay.com', y='Understock.com', 
        marker='o', s=200, 
        color='white', edgecolor='black'
    )
);
```

### Plotting with seaborn

```python
import seaborn as sns
```

```python
sns.pairplot(daily_sales_df[['iBay.com', 'Understock.com', 'Shoppe.com']]);
```

```python
grid = sns.pairplot(daily_sales_df[['iBay.com', 'Understock.com', 'Shoppe.com']]);

grid.axes
```

```python
grid = sns.pairplot(daily_sales_df[['iBay.com', 'Understock.com', 'Shoppe.com']])

first_row = grid.axes[0]
first_row[0].set_title('First')
first_row[1].set_title('Second')
first_row[2].set_title('Third');
```

### Using matplotlib, pandas and seaborn together

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(12, 6))
ax_left = fig.add_subplot(1, 2, 1) # rows, columns, position
ax_right = fig.add_subplot(1, 2, 2)
```

```python
# Anothe way
fig, (ax_left, ax_right) = plt.subplots(1, 2, figsize=(12, 6))
```

```python
fig, grid = plt.subplots(3, 3, figsize=(12, 6))
```

```python
fig, grid = plt.subplots(3, 3, figsize=(12, 6))

ax_top_left = grid[0][0]
ax_mid_center = grid[1][1]
ax_bottom_right = grid[2][2]

ax_top_left.plot([1, 2, 3], [1, 2, 3], linewidth=10, color='tomato')
ax_mid_center.scatter([1, 2, 3], [1, 2, 3], s=300, color='gold')
ax_bottom_right.bar(['a', 'b', 'c'], [1, 2, 3], color='purple');
```

```python
fig, (ax_left, ax_right) = plt.subplots(1, 2, figsize=(12, 6))

# pandas plot on the left
daily_sales_df.plot(
    ax=ax_left,
    legend='reverse'
)

# seaborn plot on the right
sns.regplot(
    ax=ax_right,
    data=daily_sales_df,
    x='iBay.com',
    y='Understock.com'
);
```
